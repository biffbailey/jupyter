## Question:
### How do the drive train on my bike work to make it go?  Why is one gear combination easier or harder to pedal or make the bike go faster or slower than some another?  How do the gear combinations compare to each other?

## Hypothesis:
### By measuring the wheel motion in response to pedal motion, observation of the physical properties of the parts and some charting and math, answers will emerge.

## Experiment Design:
### 1. Label and describe the parts.
### 2. Create lists of the possible (variable) gear selections.
### 3. Combine the variable gear selections into a list of all the possible combinations.
### 4. For each combination, measure wheel motion versus pedal motion, in revolutions and distance, feet.
### 5. Make a table of the results.
### 6. Add to the table the four basic math relationships between the wheel motion and pedal motion.
### 7. Analyse the table by charting to see which, if any, relationships explain the results.

## Variable(s):
### The input variables are:
 * Gear Combination
 * Pedal Movement, 
   + Unit: Revolutions

### The output variable is:
 * Rear Wheel Movement
  + In units: Revolutions, or
  + Distance Traveled

## Predicted Observations, Data:

```python
#Set a list variable for the Chain_Rings, Cassette_Gears and a variable for the Tire_Circumfrence
Chain_Rings = [42, 52]
Cassette_Gears = [28, 24, 21, 18, 16, 14, 12, 11]
#The bike tire is 25-520 (24x1), with approximate outer circumfrence of 1753 mm per Cateye published chart
#www.cateye.com/data/resources/Tire_size_chart_ENG.pdf
Tire_Circumfrence = 1753.0/25.4/12
#Instruct the computer to generate a list of the possible gear combinations:
#Create a list of gear combinations, and compute the expected wheel turns and wheel distance for each.
#Store the data in a dictionary with the Cassette_Gear as the key and the data list as the value.
Gear_Combinations = []
Data_Dictionary = {}
for Ring in Chain_Rings:
    for Gear in Cassette_Gears:
        Key = str(Ring)+"/"+str(Gear)
        Gear_Combinations.append(Key)
        WR = float(3*Ring)/Gear
        WD = float(WR*Tire_Circumfrence)
#Add the computed measures
        Sum = Ring + Gear
        Diff = Ring - Gear
        Prod = Ring * Gear
        Quotient = float(Ring) / Gear
        Data_Dictionary[Key]=[Ring, Gear, 3, WR, WD, Sum, Diff, Prod, Quotient]
#Print the Data
#Print header
print "Combination".center(15,'.'),\
    "Ring".center(6,'.'),\
    "Gear".center(6,"."),\
    "Pedal Rotations".center(20,'.'),\
    "Wheel Rotations".center(20,'.'),\
    "Wheel Distance".center(20,'.'),\
    "Sum".center(6,'.'),\
    "Diff".center(6,'.'),\
    "Prod".center(6,'.'),\
    "Quot".center(6,'.')
for Key in Gear_Combinations:
     print str(Key).center(15),\
        str(Data_Dictionary[Key][0]).center(6),\
        str(Data_Dictionary[Key][1]).center(6),\
        str(Data_Dictionary[Key][2]).center(20),\
        str(Data_Dictionary[Key][3]).center(20),\
        str(Data_Dictionary[Key][4]).center(20),\
        str(Data_Dictionary[Key][5]).center(6),\
        str(Data_Dictionary[Key][6]).center(6),\
        str(Data_Dictionary[Key][7]).center(6),\
        str(Data_Dictionary[Key][8]).center(6)\
        #print Gear_Combinations
#print Tire_Circumfrence," , ",type(Tire_Circumfrence)
```

## Conclusion:

## 1.  It takes a lot more effort to turn the pedals with the small gears because to turn the pedals each time the bike has to go a lot farther!
## 2.  The fastest I can go is 80 feet for 3 pedal turns or $\frac{80}{3}\approx 26$ feet per pedal turn.  That's a lot, especially if I go up a hill!
## 3.  The slowest I can go is 26 feet for 3 pedal turns or $\frac{26}{3}\approx 8$ feet per pedal turn.  This might be a lot easier to push up a hill!
## 4.  The number of wheel turns for three pedal turns $ = 3 \times \frac{Ring}{Gear}$
## 5.   So I think the number of wheel turns for one pedal turn $ = \frac{Ring}{Gear}$
